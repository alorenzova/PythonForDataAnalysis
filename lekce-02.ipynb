````python
is_magnesium = food_nutrient["name"] == "Magnesium, Mg"
is_over_limit = food_nutrient["amount"] > 100
````

````python
# dataframe[(podminka 1) & | (podminka 2)]

food_nutrient[(is_magnesium)
              & (is_over_limit)
]
````

````python
food_nutrient[(food_nutrient["name"] == "Magnesium, Mg")
              & (food_nutrient["amount"] > 100)
]
````

````python
calcium = food_nutrient[
    (food_nutrient["name"] == "Calcium, Ca")
    & (food_nutrient["amount"] > 30)
    & (food_nutrient["amount"] < 500)
]
calcium
````

````python
food_nutrient[(food_nutrient["name"] == "Calcium, Ca") & (food_nutrient["amount"] > 30) & (food_nutrient["amount"] < 500)]

food_nutrient.query('name == "Calcium, Ca" and 30 < amount < 500')

food_nutrient[(food_nutrient["name"] == "Calcium, Ca") & food_nutrient["amount"].between(30, 500, inclusive="neither")]

food_nutrient.loc[(food_nutrient["name"] == "Calcium, Ca") & (food_nutrient["amount"] > 30) & (food_nutrient["amount"] < 500)]
````

````python
# Vyfiltruj hodnoty obsahující 'calcium' bez ohledu na velikost písmen
calcium_names = food_nutrient[food_nutrient["name"].str.contains("calcium", case=False)]
calcium_names["name"].unique()
````
